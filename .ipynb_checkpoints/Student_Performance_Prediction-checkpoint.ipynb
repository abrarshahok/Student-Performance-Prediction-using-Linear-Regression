# Student Performance Prediction - Linear Regression

1. **school**: School the student attends. Example:
   - `GP`: Gabriel Pereira (a specific school).
   
2. **sex**: Gender of the student. Example:
   - `F`: Female.
   
3. **age**: Student's age (in years). Example:
   - `18`, `17`, `15`, `16`.
   
4. **address**: Type of address where the student lives. Example:
   - `U`: Urban.
   - `R`: Rural.

5. **famsize**: Size of the student's family. Example:
   - `GT3`: Family size greater than 3.
   - `LE3`: Family size less than or equal to 3.

6. **Pstatus**: Parents' cohabitation status. Example:
   - `A`: Apart (parents not living together).
   - `T`: Together (parents living together).

7. **Medu**: Mother's education level (from 0 to 4). Example:
   - `4`: Higher education (e.g., university).
   - `1`: Primary education.

8. **Fedu**: Father's education level (from 0 to 4). Example:
   - `4`: Higher education.
   - `1`: Primary education.

9. **Mjob**: Mother's job. Example:
   - `at_home`: Homemaker.
   - `teacher`: Teacher.

10. **Fjob**: Father's job. Example:
    - `services`: Working in services sector.
    - `other`: Other types of jobs.

11. **famrel**: Quality of family relationships (from 1 to 5, where 5 is the best relationship). Example:
    - `4`: Good family relationship.
    
12. **freetime**: Amount of free time after school (from 1 to 5, where 5 is a lot of free time). Example:
    - `3`: Average amount of free time.

13. **goout**: Frequency of going out with friends (from 1 to 5, where 5 is very frequent). Example:
    - `4`: Frequent outings with friends.

14. **Dalc**: Workday alcohol consumption (from 1 to 5, where 5 is very high). Example:
    - `1`: Very low alcohol consumption.

15. **Walc**: Weekend alcohol consumption (from 1 to 5, where 5 is very high). Example:
    - `1`: Low alcohol consumption.

16. **health**: Student's current health status (from 1 to 5, where 5 is excellent health). Example:
    - `3`: Average health.

17. **absences**: Number of school absences. Example:
    - `4`, `2`, `6`: Number of days absent from school.

18. **G1, G2, G3**: Grades for the first (G1), second (G2), and final (G3) periods. Example:
    - `G1`: 0, 9, 12, 14 (grades for the first term).
    - `G2`: 11, 11, 13, 14 (grades for the second term).
    - `G3`: 11, 11, 12, 14 (final grade).

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_kde(data, category, title, x_label):
    plt.figure(figsize=(12, 5))
    category_values = sorted(data[category].unique())
    colors = sns.color_palette("husl", len(category_values))

    for i, value in enumerate(category_values):
        rural_data = data.G3[(data['address'] == 'R') & (data[category] == value)]
        urban_data = data.G3[(data['address'] == 'U') & (data[category] == value)]

        # Check for variance in rural data
        if rural_data.nunique() > 1:  # More than 1 unique value
            sns.kdeplot(rural_data, 
                        fill=True, label=f'Rural Students ({category.capitalize()}: {value})', 
                        color=colors[i], warn_singular=False)

        # Check for variance in urban data
        if urban_data.nunique() > 1:  # More than 1 unique value
            sns.kdeplot(urban_data, 
                        fill=True, label=f'Urban Students ({category.capitalize()}: {value})', 
                        color=colors[i], linestyle='--', warn_singular=False)

    plt.legend(loc='upper right', fontsize='small')
    plt.title(title, fontsize=16)
    plt.xlabel(x_label, fontsize=14)
    plt.ylabel('Density', fontsize=14)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

## Preprocess Data

```python
columns_to_remove = [
    'Mjob', 'Fjob', 'famsize', 'reason',
    'Pstatus', 'paid', 'activities', 
    'nursery', 'internet', 'romantic', 
    'famrel', 'Dalc', 'Walc','Medu', 'Fedu',
    'higher'
]

df.drop(columns=columns_to_remove, inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# Converting Guardian to numeric Values, Father(1), Mother(1), Other(0)
df['guardian'] = df['guardian'].apply(lambda x: 1 if x == 'father' else (1 if x == 'mother' else 0))

# Converting Fam Support to numeric Values, Yes(1), No(0)
df['famsup'] = df['famsup'].apply(lambda x: 1 if x == 'yes' else 0)

# Converting School Support to numeric Values, Yes(1), No(0)
df['schoolsup'] = df['schoolsup'].apply(lambda x: 1 if x == 'yes' else 0)

# Converting Address to numeric Values, U(1), R(0)
df['address'] = df['address'].apply(lambda x: 1 if x == 'U' else 0)

# Converting school to numeric Values, GP(0), MS(1)
df['school'] = df['school'].apply(lambda x: 1 if x == 'MS' else 0)

# Converting Gender/sex to numeric Values, M(1), F(0)
df['sex'] = df['sex'].apply(lambda x: 1 if x == 'M' else 0)
```

## Now Prepare and Train Linear Regression Model

```python
X = df.drop(columns=['G3'])
y = df['G3'] 
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error ,r2_score, accuracy_score
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train) 
X_test_scaled = scaler.transform(X_test) 
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, y_train)
```

```python
y_pred = lr_model.predict(X_test_scaled)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
print(f"R² Score: {r2:.2f}")
```

```python
print('Original Outputs:',  list(y_test),'\n')
print('Predicted Outputs:', list(y_pred))
```
